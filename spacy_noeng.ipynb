```python
nlp = spacy.load("ru_core_news_sm")

def get_embeddings(text: str, nlp: Language) -> np.array:
    result_dict = {}
    
    doc = nlp(text=text)
    for token in doc:
        result_dict[token] = token.vector
    
    return result_dict
```

```python
word_vectors: dict = get_embeddings(text="мин вода Mattoni", nlp=nlp)
for index, value in word_vectors.items():
    print(index, value)
```
